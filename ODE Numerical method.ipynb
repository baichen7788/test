```python
from math import exp
#Analytical
f=lambda x: exp(x**2/2)

#Euler
dy=lambda x,y: x*y
x=0     #domain x from 0 to 2
xn=2
y=1   #initial y =1
h=0.05   #step size h
n=int((xn-x)/h)
print('  x \t    y(Euler) \t  y(Analytical) ')
for i in range(n):
    y=y+dy(x,y)*h
    x=x+h
    print(' {:.2f} \t\t {:.2f} \t\t {:.2f}'.format(x,y,f(x)))
    
    
#Euler is not a good method
```

```python
# RK Method
from math import exp
#Analytical
f=lambda x: exp(x**2/2)

#RK
dy=lambda x,y: x*y
x=0     #domain x from 0 to 2
xn=2
y=1   #initial y =1
h=0.05   #step size h
n=int((xn-x)/h)
print('  x \t    y(RK method) \t  y(Analytical) ')
for i in range(n):
    y=y+dy(x+h/2,y+h*dy(x,y)/2)*h
    x=x+h
    print(' {:.2f} \t\t {:.2f} \t\t {:.2f}'.format(x,y,f(x)))
```

```python
#Fourth order RK method

#Analytical
f=lambda x: exp(x**2/2)

#FOurth order RK
dy=lambda x,y: x*y
x=0     #domain x from 0 to 2
xn=2
y=1   #initial y =1
h=0.05   #step size h
n=int((xn-x)/h)
print('  x \t    y(Fourth order RK) \t  y(Analytical) ')
for i in range(n):
    k1=h*dy(x,y)
    k2=h*dy(x+h/2,y+k1/2)
    k3=h*dy(x+h/2,y+k2/2)
    k4=h*dy(x+h,y+k3)
    y=y+(k1+2*k2+2*k3+k4)/6
    x=x+h
    print(' {:.2f} \t\t {:.6f} \t\t {:.6f}'.format(x,y,f(x)))
```

```python
#Fourth order RK method (another example)
import numpy as np
import matplotlib.pyplot as plt

#Analytical solution
f=lambda x: np.exp(x-2)-3  #Must use np.exp because np.array will pass to the function

#FOurth order RK
dy=lambda x,y:y+3
x=2     #domain x from 0 to 2
xn=4
y=-2   #initial y =1
h=0.1   #step size h
n=int((xn-x)/h)
xp=np.linspace(x,xn,n+1)
#fp=f(xp)
yp=np.empty(n+1,float)
yp[0]=y
print('  x \t    y(Fourth order RK) \t  y(Analytical) ')
for i in range(1,n+1):
    k1=h*dy(x,y)
    k2=h*dy(x+h/2,y+k1/2)
    k3=h*dy(x+h/2,y+k2/2)
    k4=h*dy(x+h,y+k3)
    y=y+(k1+2*k2+2*k3+k4)/6
    yp[i]=y
    x=x+h
    print(' {:.2f} \t\t {:.6f} \t\t {:.6f}'.format(x,y,f(x)))
```

```python
#Plot RK4
plt.plot(xp,yp,'ro',xp,f(xp))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['RK4','Analytical'])
plt.show()
```

```python
#RK4 to solve High-order Plot
import numpy as np

dy=lambda x,y,u:u
du=lambda x,y,u: 4*x+10*np.sin(x)-y

#exact solution
f=lambda x:9*np.pi*np.cos(x)+7*np.sin(x)+4*x-5*x*np.cos(x)
df=lambda x:-9*np.pi*np.sin(x)+7*np.cos(x)+4-5*(np.cos(x)-x*np.sin(x))

#initial values
x=np.pi
xn=2*np.pi
y=0
u=2
h=0.1
n=int((xn-x)/h)

#Use array number to plot
xp=np.linspace(x,xn,n+1)
yp=np.empty(n+1,float)
up=np.empty(n+1,float)
yp[0]=0
up[0]=2
print('x \t\t y\'(RK4) \t y(Rk4) \t y\'(Exact) \t y(Exact)')
for i in range(1,n+1):
    L1=h*du(x,y,u)
    K1=h*dy(x,y,u)
    
    L2=h*du(x+h/2,y+K1/2,u+L1/2)
    K2=h*dy(x+h/2,y+K1/2,u+L1/2)
    
    L3=h*du(x+h/2,y+K2/2,u+L2/2)
    K3=h*dy(x+h/2,y+K2/2,u+L2/2)
    
    L4=h*du(x+h,y+K3,u+L3)
    K4=h*dy(x+h,y+K3,u+L3)
    
    u= u+(L1+2*L2+2*L3+L4)/6
    up[i]=u
    y=y+(K1+2*K2+2*K3+K4)/6
    yp[i]=y
    x=x+h
    
    print('{:.5f} \t {:.5f} \t {:.5f} \t {:.5f} \t {:.5f}'.format(x,u,y,df(x),f(x)))
plt.plot(xp,yp,'ro',label='y RK(4)')
plt.plot(xp,up,'bs',label='y\' (RK4)')
plt.plot(xp,f(xp),'k-',lw=2,label='y(Exact)')
plt.plot(xp,df(xp),'k--',lw=2,label='y\'(Exact)')
plt.legend()
plt.grid()
plt.show()
```

```python
#Solve ODE using Scipy
from scipy.integrate import odeint
from numpy import linspace,arange

#1st ODE equation
dy=lambda y,x: x*y
y0=1
x=linspace(0,2,5)
y=odeint(dy,y0,x)
print(y)
```

```python
#2nd ODE equation
dy=lambda y,x:y+3
y0=-2
x=linspace(2,4,21)
y=odeint(dy,y0,x)
print(y)
```
